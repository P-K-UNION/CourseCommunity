<p><span style="font-size: 18px;">

### 一、资本资产定价模型
<p><span style="font-size: 18px;">
根据$CAPM$理论，资产的预期超额收益可以由下面这个一元线性模型决定：
    
<p><span style="font-size: 18px;">
    $$E[R_i]-R_f=\beta_i(E[R_M]-R_f), where \beta_i=cov(R_i,R_M)/var(R_M)$$

<p><span style="font-size: 18px;">
其中$E[]$是期望符号，$R_i$是某资产$i$的收益率，$R_f$是无风险收益率，$R_M$是市场组合的收益率，
$\beta_i$代表了该资产对市场收益的没敏感程度，也即对市场风险的暴露程度。$CAPM$是最简单的线性因子模型，指出资产的预期超额收益率取决于资产对市场风险的暴露大小，市场组合也被称为市场因子。

### 二、套利定价理论
<p><span style="font-size: 18px;">
$CAPM$模型为后续大量线性多因子定价模型的研究拉开序幕，人们发现不同资产的收益率并非由单一的市场因子决定，而是同时受到多种不同因子影响（例如企业的财务业绩等等），这也就和$CAPM$相矛盾。从而$Ross(1976)$很自然提出了著名的套利定价理论$APT$，其假定资产的预期超额收益由一下多元线性模型决定：

<p><span style="font-size: 18px;">
    $E[R_i^e]=\alpha_i+\beta_i\lambda+\epsilon_i$

<p><span style="font-size: 18px;">
其中$E[R_i^e]$指的是资产预期超额收益,$E[R_i^e]$的表述方便了资金中性投资组合，更具有普遍性。$\beta_i$是资产i的因子暴露或因子载荷，$\lambda$是因子预期收益也即因子溢价。$\alpha_i$往往是实际预期收益率和多因子模型隐含的预期收益率之间的定价误差。

<p><span style="font-size: 18px;">
定价误差产生的原因可能是：
（1）模型设定错误，即有遗漏的因子。
（2）模型没有问题。若$\alpha_i$并不显著，则定价误差可能是运气；若$\alpha_i$显著，则代表了一个套利获得超额收益的机会。

### 三、异象检验局限性
<p><span style="font-size: 18px;">
（一）正面结果的发表偏见。在学术研究中，积极的、显著的结果比负面的、无显著性结果更容易被发表。这种偏见可能导致异象检验的结果过于乐观。
<p><span style="font-size: 18px;">
（二）统计阈值。当阈值设置过低时，更多的异象可能被误认为是显著的，而实际可能并没有经济意义。
<p><span style="font-size: 18px;">
（三）异象公开后的消失。一旦某个异象被广泛报道或公开，市场参与者可能会迅速对其做出反应，从而导致这一异象的效果消失。
<p><span style="font-size: 18px;">
（四）羊群效应。投资者倾向于跟随大多数人的行为，而不是基于自己的独立分析做出决策。当一个异象被广泛接受和使用时，更多的投资者可能会跟随这一异象的投资策略，从而影响市场价格并可能改变异象的表现。

（三）预测性回归、排序和p值问题

<p><span style="font-size: 18px;">
（1）预测性回归，假设以下简单的形式：
<p><span style="font-size: 18px;">
$$r= a+bx+e$$

<p><span style="font-size: 18px;">
其中，向量$r$是所有股票的收益构成，$x$是滞后变量，因而这个回归具有预测性。若在给定阈值下，估计值$ \hat{b} $是显著的，则$x$在预测收益方面可能有效。当然，$ \hat{b} $提供的只是过往预测能力的信息，预测未来往往出错。

<p><span style="font-size: 18px;">
（2）统计检验，也被用于对投资组合进行分析。假设两个极端的投资组合（例如大市值公司和小市值公司），可以采用t检验：

<p><span style="font-size: 18px;">
$$t= \sqrt{T}\frac{m_{r_+}-m_{r_-}}{\sigma_{{r_+}-{r_-}}}$$


<p><span style="font-size: 18px;">
其中，$T$是观测值的数量，$m_{r_+}$和$m_{r_-}$是收益率平均值，$\sigma_{{r_+}-{r_-}}$是两个收益序列差值的标准差，即多空组合的波动率。换言之，这个统计数字可以视为缩放的夏普比率，可以反过来用于计算$p$值评估异象的稳定性。

<p><span style="font-size: 18px;">
（3）$p$值的误解
<p><span style="font-size: 18px;">
传统的p值通常用来衡量假设检验中观测结果相对于原假设的极端性。问题在于，人们常常误解或过度依赖p值，将它视为显著性和效应真实性的直接衡量标准。而传统p值未考虑先验知识，只基于数据得出结论，忽略了研究者对因子效应已有的先验认知或历史信息；低p值并不直接代表结果的效应强度或具有经济意义，投资者常常过度解释“显著性”，忽视其实际经济含义；因子投资中会同时检验多个策略或因子，传统p值没有很好地处理多重比较问题，可能导致假阳性（Type I 错误）。

<p><span style="font-size: 18px;">
$p$值常常被理解为给定数据的前提下假设为真（即异象存在）的概率。事实却是，假设为真的前提下观察到数据的条件概率。

<p><span style="font-size: 18px;">
$$p值=P[D|H]$$
<p><span style="font-size: 18px;">
$$目标概率=P[H|D]=\frac{P[D|H]}{P[D]}×P[H]$$

<p><span style="font-size: 18px;">
贝叶斯化的p值（或贝叶斯p值）确实可以帮助解决传统p值带来的误解。
$$Bayesianized-p-value=e^{-t^2/2}×\frac{prior}{1+e^{-t^2/2}×prior}$$

<p><span style="font-size: 18px;">
其中t是回归得到的t统计量，prior是分析师对假设为真的先验机会比。贝叶斯方法结合了先验分布（代表已有知识或假设）与数据后验分布，能够提供更全面的结论。贝叶斯p值直接给出假设为真的概率，而不是传统的“假设不成立时观测结果的概率”，使得解释更加直观。贝叶斯框架下，可以更合理地解释置信区间和效应估计，减少投资者对单一p值的过度依赖。缺点是先验机会比对于计算影响很大，却完全由使用者经验决定。

（三）Fama-MacBeth回归法

<p><span style="font-size: 18px;">
两步回归法估计，另一种检验异象的方法。

<p><span style="font-size: 18px;">
第一步对所有股票在时间序列上逐个回归:
<p><span style="font-size: 18px;">
$$E[R_i^e]=\alpha_i+\beta_i\lambda+\epsilon_i$$

<p><span style="font-size: 18px;">
第二步将得出的$\hat{\beta_i}$代入第二步回归中:

<p><span style="font-size: 18px;">
$$r_{t,i}=\gamma_{t,0}+\sum_{i=1}^{N}\gamma_{t,k}\hat{\beta_{i,k}}+\epsilon_{t,i}$$
<p><span style="font-size: 18px;">
$\hat{\gamma_{t,k}}$是t时刻下因子$k$溢价的估计值。在对$\epsilon$分布的合理假设下，可以通过统计检验来确定溢价是否显著。通常，时间加权溢价$\hat{\gamma_k}=\frac{1}{T}\sum_{i=1}^{T}\gamma_{t,k}$的统计量为：
<p><span style="font-size: 18px;">
$$t_k=\frac{\hat{\gamma_{k}}}{\hat{\sigma_k}/\sqrt{T}}$$
<p><span style="font-size: 18px;">
用于评估在纯高斯假设下，因子是否显著。由于收益率序列未能找到，故仅将代码列在下方。

<p><span style="font-size: 18px;">
（四）因子竞争

<p><span style="font-size: 18px;">
因子主要用于解释股票收益率的界面差异（即相同时间节点上，不同股票之间的收益率差异）。从理论和实际上讲，我们都希望避免冗余的因子。冗余意味着共线性（高绝对相关性），会对估计结果产生干扰。此外，收益率对高度共线性因子的正向和负向的高暴露值以虚假方式相互补偿，故当资产管理人将收益分解到各个因子，因子间的共线性导致的风险暴露回归结果是难以解释的。换句话说，高度共线性意味着它们的变动趋势非常相似，难以单独区分它们各自对资产回报的影响。同时，正向暴露和负向暴露的因子对收益的影响可能会相互抵消。
<p><span style="font-size: 18px;">
可以将每个因子与其他因子进行回归以识别冗余的因子：
<p><span style="font-size: 18px;">
$$f_{t,k}=a_k+\sum_{j≠k}\delta_{k,j}f_{t,j}+\epsilon_{t,k}$$
<p><span style="font-size: 18px;">
$a_k$是一个有意义的指标，如果其显著不等于0，那么其他因子不能完全解释该因子的收益。否则，其是冗余的。


```python
import numpy as np
import statsmodels.api as sm

# 初始化一个列表，用于存储每次回归的结果
df_res_full = []

# 循环遍历因子列表（共5个因子：MKT_RF, SMB, HML, RMW, CMA）
for i in range(0, 5):
    # 因子名称列表
    factors_list_full = ["MKT_RF", "SMB", "HML", "RMW", "CMA"]
    
    # 创建一个临时因子列表，最初包含所有因子
    factors_list_tmp = factors_list_full.copy()
    
    # Y为被解释变量（即当前循环的因子）
    Y = FF_factors[factors_list_full[i]]
    
    # 从临时因子列表中移除当前因子，用于解释该因子的其它因子作为自变量
    factors_list_tmp.remove(factors_list_full[i])
    
    # 使用剩下的因子作为自变量，构造数据
    data = FF_factors[factors_list_tmp]
    
    # 使用statsmodels库进行OLS（普通最小二乘）回归，add_constant是添加常数项
    results = sm.OLS(endog=Y, exog=sm.add_constant(data)).fit()
    
    # 提取回归系数
    results_param = results.params
    
    # 将回归结果保存到一个DataFrame中
    reg_result_tmp = pd.DataFrame(results_param)
    
    # 增加一个列，记录当前被回归的因子名称
    reg_result_tmp['factor_mnemo'] = Y.name
    
    # 增加一个列，保存每个回归系数的p值
    reg_result_tmp['pvalue'] = results.pvalues
    
    # 将临时结果DataFrame添加到总的结果列表中
    df_res_full.append(reg_result_tmp)

# 将所有回归结果合并为一个DataFrame
df_res_full = pd.concat(df_res_full)

# 重置索引，确保DataFrame索引的正确性
df_res_full.reset_index(inplace=True)

# 将系数列重命名为“coeff”
df_res_full.rename(columns={0: "coeff"}, inplace=True)
```

```python
import numpy as np
import pandas as pd

df_significance = pd.DataFrame(df_res_full)

# 构建条件列表，筛选 p-value 以分配星级符号
conditions = [
    (df_significance['pvalue'] > 0) & (df_significance['pvalue'] < 0.001),
    (df_significance['pvalue'] > 0.001) & (df_significance['pvalue'] < 0.01),
    (df_significance['pvalue'] > 0.01) & (df_significance['pvalue'] < 0.05),
    (df_significance['pvalue'] > 0.05)
]

# 对应的星级符号
valuest = ['(***)', '(**)', '(*)', 'na']

# 创建一个新列，标注星级
df_significance['stars'] = np.select(conditions, valuest)

# 创建空的矩阵框架
row_names = ["MKT_RF", "SMB", "HML", "RMW", "CMA"]
col_names = [ "MKT_RF", "SMB", "HML", "RMW", "CMA"]
result_matrix = pd.DataFrame(index=row_names, columns=col_names)

# 填充矩阵
for i in range(len(df_significance)):
    row = df_significance.loc[i, 'factor_mnemo']  # 行名
    col = df_significance.loc[i, 'index']  # 列名
    value = df_significance.loc[i, 'coeff']  # 系数
    star = df_significance.loc[i, 'stars']  # 星级
    # 组合系数和星级
    result_matrix.loc[row, col] = f"{value:.3f} {star}"

# 打印最终结果
result_matrix
```

<p><span style="font-size: 18px;">
从表格可以看到，当其他四个因子存在于定价模型中，const几乎为0，HML即价值因子是冗余的。

宏观层面上，研究人员试图找到基于实证观察，哪些因子组合最有效，例如市场因子、规模因子和价值因子三因子模型（Fama & French,1993）多大程度优于新增盈利因子（RMW，robust minus weak）和投资因子（CMA，conservative minus aggressive）的五因子模型（Fama & French,2015）。研究人员引入新的p值计算方法，比较模型通过0-α检验的相对可能性。其他研究还有包含贝叶斯方法，和从交易成本角度进行模型比较。

<p><span style="font-size: 18px;">
因子最优数量众说纷纭，传统文献关注3~5个，最新研究主张需要使用至少15个，另一些主张少量隐性因子即足够。一些研究发现有助于解释收益率截面数据的特征数量是随时间变化的。

<p><span style="font-size: 18px;">
（五）进阶

<p><span style="font-size: 18px;">
随着研究深入，被发现的因子越来越多，其在资产管理领域也越来越重要，促进研究设计更精密方法来管理“因子动物园”，检测虚假异象和比较不同的资产定价模型。

Feng、Giglio 和 Xiu （2020） 将 LASSO 选择与 Fama-MacBeth 回归相结合，以测试新的因子模型是否值得。他们量化了在一组预定义因子中添加一个新因子的收益，并表明 2010 年十年发表的论文中报告的许多因子并没有增加太多增量价值;

C. Harvey 和 Liu （2019） （以类似的方式） 对正交因子使用 bootstrap。他们指出，预测因子之间的相关性是一个主要问题，他们的方法旨在解决这个问题。他们冗长的过程旨在测试候选变量的最大额外贡献是否重要;

Fama 和 French （2018） 通过最大夏普比率的平方来比较资产定价模型;

Giglio 和 Xiu （2019） 使用基于主成分分析的三遍法估计因素风险溢价;

Pukthuanthong、Roll 和 Subrahmanyam （2018） 通过结合主成分分析和 Fama 和 MacBeth （1973） 回归来解开定价和非定价因GospGospodinov、Kan 和 RGospodinov、Kan 和 RGospodinov、Kan 和 RGospodinov、Kan 和 Robotti （2019） 警告要注意模型设定错误（假因子被包括在回归方程的解释变量中）。而交易因子（相对于宏观经济因素）似乎更有可能产生稳健收益（  Bryzgalova （2019））。

<p><span style="font-size: 18px;">
五、因子和特征之争

<p><span style="font-size: 18px;">
在学术文献中，公司收益由宏观经济因子（如市场、规模、账面市值比）还是公司特性（如公司基本财务特征）来解释，存在争议。换句话说，解释收益时，公司的核心特征和宏观经济因子哪种影响更明显、更重要？

主要观点包括：

特性 vs. 因子暴露：Lakonishok, Shleifer 和 Vishny (1994) 提出，价值溢价的部分来源可能是投资者对过去盈利增长率的错误推断，过度乐观的投资者更倾向于收益较高的公司，因此未来的回报是由公司的核心财务特征驱动的，而非宏观经济因子。

Daniel 和 Titman (1997) 及后续研究表明，具有高账面市值比或小市值的公司，即使在 HML 或 SMB 因子上呈现负相关，也能获得较高的平均回报。因此，公司特性在解释收益时更为重要，而不是因子暴露。

后续研究：如 Goyal (2012)、Chordia、Goyal 和 Shanken (2019) 等人的研究进一步探讨了特性和因子在收益预测中的作用。研究发现，公司的特性比因子负荷能更好地解释预期收益的变化。

最新方法：近年来，学者们采用了诸如 LASSO 和 ElasticNet 等方法对特性进行回归分析，旨在识别对收益预测有用的特性，并通过严格的回归模型避免包含无关的特性。

市场需求模型：Koijen 和 Yogo (2019) 提出了一种基于投资者需求的模型，表明投资者的组合偏好基于公司特性，而价格的变化与特性直接相关，而不是与因子相关。

总结而言，文章讨论了关于公司特性与宏观经济因子在收益预测中的重要性争论，越来越多的证据表明公司特性在解释和预测资产收益方面占据更重要的地位。

六、热点
<p><span style="font-size: 18px;">（一）因子动量
<p><span style="font-size: 18px;">
（1）因子动量
因子收益率在时间序列上存在动量效应。因子收益率的自相关模式在统计上显著。有些研究表明，其只对少数因子有稳健性。对于因子动量，对其来源的研究将其分解为因子择时组合和静态组合两个部分，前者试图从因子收益的序列相关性获益（即通过分析因子收益的时间序列相关性来决定在不同时间点对某个因子进行加仓或减仓），后者试图捕捉因子溢价（即捕捉因子的长期溢价）。研究表明，静态组合解释了因子动量收益率较大部分。因子的自相关函数（ACF）计算公式如下图：
<p><span style="font-size: 18px;">
$$ACF_{k}(x_t)=E[(x_t-\bar{x})(x_{t+k}-\bar{x})]$$

```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm

fig, ax = plt.subplots(2,2,figsize=(20,20),sharex='all', sharey='all')
sm.graphics.tsa.plot_acf(FF_factors.RMW, lags=10, ax=ax[0,0],title='RMW') # ACF回归
sm.graphics.tsa.plot_acf(FF_factors.CMA, lags=10, ax=ax[1,0],title='CMA')
sm.graphics.tsa.plot_acf(FF_factors.SMB, lags=10, ax=ax[0,1],title='SMB')
sm.graphics.tsa.plot_acf(FF_factors.HML, lags=10, ax=ax[1,1],title='HML')
plt.show()
```

<p><span style="font-size: 18px;">
可见四个因子中，只有规模因子SMB的一阶自相关系数不显著。

<p><span style="font-size: 18px;">
（2）因子动量
有大量证据表明因子溢价的时变性，对于因子何时表现良好或糟糕的预测应运而生。当然，择时有效性的证据充满分歧，对于使用哪些预测变量，业界也没有共识。有采用一般宏观经济指标，有采用股票发行与回购量，也有采用总体基本面数据。在基于机器学习的因子投资中，有可能通过将公司的特定属性与大规模的经济数据结合，得到更准确结果。

新世纪以来，人们对道德金融产品需求急剧上升，催生了专门用于社会责任投资的基金，其加速发展驱动了关于ESG（环境、社会、治理）标准相关的公司特征是否可以被定价。当然ESG投资的有效性仍众说纷纭。但ESG标准或可整合到机器学习中。

<p><span style="font-size: 18px;">
（二）与机器学习的联系

<p><span style="font-size: 18px;">
鉴于可用数据的指数级增长，基金经理总是试图从公司的大量特征中推断公司未来收益。这里的特征不仅包括会计比率等经典数据，还包括情绪等另类数据。机器学习恰好契合此类项目。如果基本面数据（会计比率、收益、相对估值等）有助于预测收益，那么一个细化的方法就是先预测这些基本面数据，获得信息优势。早期的一些旨在用公司特征解释和预测收益的研究，虽然没有结合机器学习模型，但其实和机器学习存在一定联系。一般表达式如下，在时间T内，投资者的目标优化函数为：

<p><span style="font-size: 18px;">
$$\max_{\theta_T}E_T[u(r_{p,T+1})]=\max_{\theta_T}E_T[u((\bar{w_T}+x_T\theta_T)'r_{T+1})]$$

<p><span style="font-size: 18px;">
其中u是效用函数，$r_{p,T+1}=(\bar{w_T}+x_T\theta_T)'r_{T+1}$是投资组合的收益率，其定义为基准$\bar{w_T}$加上由$x_T\theta_T$定义的偏离。此外，还可以增加一些约束条件，如限制杠杆等。

<p><span style="font-size: 18px;">
在实践中，向量$\theta_T$必须使用过去的$（T-\tau 到 T-1）$数据来估计，投资者的目标函数是：
<p><span style="font-size: 18px;">
$$\max_{\theta_T}\frac{1}{\tau}\sum^{T-1}_{t=T-\tau}[u(\sum^{N_T}_{i=1}(\bar{w_{i,t}}+\theta_T'x_{i,t})r_{i,t+1})]$$

<p><span style="font-size: 18px;">
对于一个样本大小为$\tau$的样本，$N_t$表示资产数量。方程即为一项学习任务，选择使平均收益率最大化的参数。
<p><span style="font-size: 18px;">
因子投资与资产定价模型之间第一个明显联系时平均收益预测，一般方程式如下：
<p><span style="font-size: 18px;">
$$r_{t+1,n}=g(x_{t,n})+\epsilon_{t+1}$$
<p><span style="font-size: 18px;">
与因子模型的第一个明显区别在于非线性函数g的引入。第二个区别在于时间指数的转变：用同期因子来解释资产收益没有用，因为因子溢价事先不知道。因而要在模型中获得价值，在观测和实际收益需要有一个时间间隔。一旦模型$\hat{g}$被估计出来，未来收益的预测便可以作为投资组合权重的信号。
<p><span style="font-size: 18px;">
其他机器学习工具也可以用来估计资产定价模型，不胜枚举，故略去，待后续补充。
